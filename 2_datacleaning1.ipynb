MULTIPLE IMPUTATION BY CHAINED EQUATIONS TO FILL MISSING VALUES IN VALUE CONFIGURATION

```python
#need to enable iterative imputer explicity since its still experimental
from sklearn.experimental import enable_iterative_imputer    # Enables use of IterativeImputer
from sklearn.impute import IterativeImputer                   # Import the actual imputer
```

```python
# What is MICE?
#MICE is a sophisticated multivariate imputation method that:
#Fills in missing values iteratively, using regression models based on other features.
#Performs better than simple imputation methods like mean, median, or mode when data is not missing completely at random.
#How it works:
#Each feature with missing values is regressed on other features.
#Missing values are predicted based on this regression.
#This is repeated in a round-robin fashion until convergence.
#This results in more statistically sound imputations, especially for complex datasets.
```

```python
#The IterativeImputer is still considered experimental in scikit-learn (even in newer versions), so to use it, you must explicitly enable it.
#Without the first line, you'll get an ImportError.
```

```python
| Use Case                            | Why MICE Helps                                              |
| ----------------------------------- | ----------------------------------------------------------- |
| Missing values in multiple features | Uses correlations between features to fill in intelligently |
| Time-series or correlated variables | Preserves dependencies across columns                       |
| Complex, non-linear relationships   | Better than mean/median imputation                          |
| Want better model performance       | Imputed data is more realistic, so models generalize better |
```

```python
#Define Imputer
imputer = IterativeImputer(random_state = 100, max_iter=10)
```

```python
#IterativeImputer(...)
#This uses Multiple Imputation by Chained Equations (MICE) to estimate missing values.
#Parameters:random_state=100
#Ensures reproducibility of results.
#Controls the randomness involved in imputation.
#Same input will always give the same output if you reuse this random_state.
#max_iter=10 = Number of maximum iterations the imputer will run to refine the imputed values.
#In each iteration, the imputer updates its guesses of the missing values.
#10 is usually a good default for convergence.

#When a column (say, A) has missing values, IterativeImputer treats A as the target variable, and the rest of the columns (B, C, etc.) 
#as the predictor variables (features). It trains a regression model using the available (non-missing) data and uses that model to predict the missing values 
#in A."""
```

```python
df1_train = df1.loc[:,['reordered','order_dow', 'order_hour_of_day' ,'days_since_prior_order']]   # df1.loc[:, [...]]
                                                                                                 #This uses the .loc[] accessor to select specific columns from the DataFrame df1.
                                                                                                 # : means select all rows.
#['reordered', 'order_dow', 'order_hour_of_day', 'days_since_prior_order'] specifies the subset of columns to include.
                                                                                                    #.sample(3)
                                                                                                    # This returns 3 random rows from df1_train.
                                                                                                    # Used for quick inspection or debugging to see sample values.
df1_train.sample(3)                                                                          # The result is stored in a new DataFrame df1_train.
```

```python
df1_train['days_since_prior_order'].value_counts(dropna=False)
```

```python
#fit on the dataset           Step 1: imputer.fit(df1_train)
                            #The IterativeImputer learns the relationships between the columns in df1_train.
                            #It analyzes observed data to build regression models that will be used to predict
                            # missing values for columns that have any missing data.
imputer.fit(df1_train)
```

```python
df_imputed = imputer.transform(df1_train)   #Step 2: df_imputed = imputer.transform(df1_train)
                                            #Now that the imputer has learned the data patterns, .transform() 
                                           #fills the missing values in df1_train using those learned regression models.
                                           #It returns a NumPy array (df_imputed) where all missing values are replaced by the imputed estimates.
                                           #The array has the same shape as df1_train.
df_imputed[:10]
```

```python
df.loc[:, ['reordered','order_dow', 'order_hour_of_day' ,'days_since_prior_order']] = df_imputed  

#This updates the original DataFrame df for the specified columns with the imputed values.
#Since df_imputed is a NumPy array and df.loc[...] selects those columns, 
#this line replaces the original columns (which may have missing values) with the fully imputed data.
#After this step, df no longer has missing values in these columns (assuming imputer filled them all).
```

```python
df['days_since_prior_order'].value_counts(dropna=False)
```

```python
#| Step            | What it does                                  | Result                                       |
| --------------- | --------------------------------------------- | -------------------------------------------- |
| `.fit()`        | Learns regression models for imputation       | Trained imputer models                       |
| `.transform()`  | Applies learned models to fill missing values | Complete dataset with no missing values      |
| `[:10]`         | Displays first 10 rows of imputed data        | Quick preview of imputed values              |
| `df.loc[...] =` | Updates original DataFrame with imputed data  | Original DataFrame's missing values replaced |
```
