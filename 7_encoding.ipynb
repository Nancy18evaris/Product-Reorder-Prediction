🔹 Step 1: Import necessary tools
OneHotEncoder: Transforms categorical (nominal) features into binary (0/1) columns.
pickle: Used to save the encoder for future use (e.g., during model deployment or inference).

🔹 Step 2: Create and configure the encoder
handle_unknown='ignore': Ignores unseen categories at transform time instead of throwing an error.
sparse_output=False: Returns a dense NumPy array (not a sparse matrix) when transforming.

🔹 Step 3: Fit and transform your nominal (categorical) columns
Learns the unique categories from each nominal column in df[nominal_cols].
Transforms each row into a one-hot encoded vector (0s and 1s).
Result: A NumPy array where each column corresponds to a unique category across all nominal_cols.


🔹 Step 4: Convert encoded array to a DataFrame
Converts the NumPy array into a Pandas DataFrame.
ohe.get_feature_names_out(nominal_cols) generates column names in the format:

🔹 Step 5: Save the fitted encoder   
Saves the fitted encoder object (ohe) to a file named 'onehot_encoder.pkl'.
This is useful to:
Apply the same encoding to test data or new data later.
Maintain consistency between training and deployment.

✅ encoded_df: Final Output
encoded_df is a clean, model-ready DataFrame with binary indicator columns.
Each row is a one-hot encoded version of the original nominal_cols.

```python
from sklearn.preprocessing import OneHotEncoder
import pickle

ohe = OneHotEncoder()
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoded_array = ohe.fit_transform(df[nominal_cols])




encoded_df = pd.DataFrame(encoded_array, columns=ohe.get_feature_names_out(nominal_cols))


# Save the fitted OneHotEncoder to a pickle file
with open('onehot_encoder.pkl', 'wb') as f:
    pickle.dump(ohe, f)

encoded_df
```

Normalizing selected numeric columns using MinMax Scaling (i.e., scaling values to a [0, 1] range).

Saving the trained MinMaxScaler to a .pkl file so you can apply the same transformation to new data later.
🔹 Step 1: Import libraries
MinMaxScaler: Scales numerical features to a specified range (default: 0 to 1).
pickle: Saves Python objects (like the fitted scaler) for future reuse.
🔹 Step 2: Define columns to normalize
These are the columns in your DataFrame df that you're scaling.
These are likely numeric features relevant to purchase behavior and sales.
🔹 Step 3: Initialize the scaler-scaler = MinMaxScaler()
Initializes a MinMaxScaler that will scale each column to [0, 1]
🔹 Step 4: Create a copy and scale
 fit_transform():
Learns the min and max values for each specified column.
Applies the scaling transformation to each value.
The scaled values replace the originals in the df_normalized DataFrame.
🔹 Step 5: Save the fitted scaler in a pickle file 
This saves the scaler object (with learned min and max) to a file.
Useful when you want to apply the same transformation to test or live data during model inference.
🔹 Step 6: Preview the result
Displays the first 5 rows of the normalized DataFrame.

✅ Result:
df_normalized has the same structure as df, but with the selected columns scaled between 0 and 1.
Good for:
Machine learning models (many models perform better with normalized input)
Consistency across different feature scales
Visualization and clustering

```python
from sklearn.preprocessing import MinMaxScaler
import pickle

# Columns to normalize
cols_to_normalize = ['add_to_cart_order',  'order_dow', 'order_hour_of_day','days_since_prior_order','product_reorder_ratio', 'total_sales', 'weekday_sales','weekend_sales']
                 

scaler = MinMaxScaler()

# Fit and transform specified columns and create a new DataFrame
df_normalized = df.copy()  # Optional: copy the original to preserve it

df_normalized[cols_to_normalize] = scaler.fit_transform(df[cols_to_normalize])
# Save the scaler as a pickle file
with open('minmax_scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

print(df_normalized.head())
```

```python
df_final = pd.concat([encoded_df, df_normalized], axis=1)
df_final
```
