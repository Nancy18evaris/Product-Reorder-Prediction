```python
import pandas as pd
import numpy as py
```

```python
aisles_df = pd.read_csv(r"F:\FINAL\DL_FINAL\New folder\aisles.csv")
aisles_df
```

```python
products_df = pd.read_csv(r"F:\FINAL\DL_FINAL\New folder\products.csv")
products_df
```

```python
aisles_product_df = pd.merge(aisles_df, products_df, on='aisle_id')
aisles_product_df
```

```python
departments_df = pd.read_csv(r"F:\FINAL\DL_FINAL\New folder\departments.csv")
departments_df
```

```python
aisles_product_departments_df1 = pd.merge(aisles_product_df, departments_df, on='department_id')
aisles_product_departments_df1
```

```python
order_products_prior_df = pd.read_csv(r"F:\FINAL\DL_FINAL\New folder\order_products__prior.csv")
order_products_prior_df
```

```python
order_products_train_df = pd.read_csv(r"F:\FINAL\DL_FINAL\New folder\order_products__train.csv")
order_products_train_df
```

```python
order_products_train_prior_df = pd.concat([order_products_prior_df, order_products_train_df])
print(order_products_train_prior_df.shape)
order_products_train_prior_df.head(5)    
```

```python
orders_df = pd.read_csv(r"F:\FINAL\DL_FINAL\New folder\orders.csv")
orders_df
```

```python
order_products_train_prior_orders_df2 = pd.merge(order_products_train_prior_df, orders_df, on='order_id')
order_products_train_prior_orders_df2
```

```python
df = pd.merge(aisles_product_departments_df1, order_products_train_prior_orders_df2)
pd.set_option("display.max_rows",None)
pd.set_option("display.max_columns",None)
df.sample(5)
```

```python
df.to_csv("after_merged.csv") # converting df to csv
```

```python
df['product_name'] = df['product_name'].astype(str).str.strip().str.lower()   #🔹 .astype(str)
                                                                              #This converts all values in the product_name column to string.
                                                                              #It's useful because the column might contain non-string values (like numbers or NaN), 
                                                                              # which could otherwise cause errors when applying string operations.
                                                                              #🔹 .str.strip()
                                                                              #This removes any leading or trailing whitespace from each string. " Apple " → "Apple"
                                                                              #🔹 .str.lower()
                                                                              # This converts all strings to lowercase."Apple" → "apple"
                                                                             

                                                                     
```

```python
df['product_name'].unique()  #returns a NumPy array of the unique values found in the product_name column of the DataFrame df.
```
